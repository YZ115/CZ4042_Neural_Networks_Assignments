```python
#
# Project 1, source code for part A question 2
#

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
import time
```

```python
# scale data (set range min and max of 0 and 1 respectively)
def scale(X, X_min, X_max):
    return (X - X_min)/(X_max-X_min)

# Default values
NUM_CLASSES = 3
NUM_FEATURES = 21

epochs = 1000
batch_size = 64
num_neurons = 10
seed = 10
decay_beta = 0.000001
learning_rate = 0.01
hidden_layer = 10

np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#read train data
def get_data():
    train_input = np.genfromtxt('ctg_data_cleaned.csv', delimiter= ',')
    trainX, train_Y = train_input[1:, :21], train_input[1:,-1].astype(int)
    trainX = scale(trainX, np.min(trainX, axis=0), np.max(trainX, axis=0))

    trainY = np.zeros((train_Y.shape[0], NUM_CLASSES))
    trainY[np.arange(train_Y.shape[0]), train_Y-1] = 1 
    
    return trainX, trainY

#Split train and test data (70:30)
def split_data(trainX, trainY):
    train_X, test_X, train_Y, test_Y = train_test_split(trainX, trainY, test_size=0.3)
    print("TrainX: ",train_X.shape)
    print("TestX: ",test_X.shape)
    print("TrainY: ",train_Y.shape)
    print("TestY: ",test_Y.shape)
    return train_X, test_X, train_Y, test_Y
```

```python
# Generate model
def get_model():
    model = keras.Sequential([
        keras.layers.Dense(hidden_layer, use_bias = True, input_shape = (NUM_FEATURES,), activation='relu', 
                           kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
        keras.layers.Dense(NUM_CLASSES, use_bias=True, activation='softmax')
    ])

    GD_opt = keras.optimizers.SGD(learning_rate=learning_rate)

    model.compile(optimizer=GD_opt,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
# Train model
def fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model):
    timeTaken=timeCallback()
    history = model.fit(train_X, train_Y,
                        epochs=epochs,
                        callbacks = [timeTaken],
                        verbose = 2,
                        batch_size=batch_size,
                        validation_data = (test_X, test_Y),
                        shuffle=True)
    timeTaken = np.mean(timeTaken.logs)
    #plt.title('Question 1A Model Accuracy')
    #plt.xlabel('No. epoch')
    #plt.ylabel('Accuracy')
    #plt.legend(['train data', 'test data'], loc='lower right')
    #plt.plot(range(epochs),history.history['accuracy'])
    #plt.plot(range(epochs),history.history['val_accuracy'])
    return history, timeTaken
```

```python
# Plot train and test accuracy against number of epochs given batch size
def plot_accuracy(acc, test_acc, epochs, batch_size):
    plt.clf()
    plt.plot(range(epochs), accuracy)
    plt.plot(range(epochs), test_accuracy)
    plt.xlabel('No. epochs')
    plt.ylabel('Accuracy')
    plt.legend(['train data', 'test data'], loc='lower right')
    plt.title(batch_size)
    
#plot_accuracy(accuracy,val_acc,100,64)
```

```python
# Function for timing duration per epoch
class timeCallback(tf.keras.callbacks.Callback):
    
    def on_train_begin(self, logs={}):
        self.logs=[]

    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_start = time.time()

    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(time.time() - self.epoch_start)
```

```python
# Implementation of k-fold cross validation
def k_fold_CV(k, dataX, dataY, epochs, batch_size):
    interval_length = len(dataX) // k
    k_accuracy = []
    k_accuracy_test = []
    k_loss = []
    k_loss_test = []
    time_log = []
    for i in range(k):
        print("fold number: ",i)
        start_time = time.time()
        
        # get train and test data
        train_X = np.concatenate([dataX[:i * interval_length],dataX[(i + 1) * interval_length:]],axis=0)
        train_Y = np.concatenate([dataY[:i * interval_length],dataY[(i + 1) * interval_length:]],axis=0)
        
        test_X = dataX[i * interval_length: (i + 1) * interval_length]
        test_Y = dataY[i * interval_length: (i + 1) * interval_length]
        
        model = get_model()
        
        history, timeTaken = fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model)
        
        time_log.append(timeTaken)
        k_accuracy.append(history.history['accuracy'])
        k_accuracy_test.append(history.history['val_accuracy'])
        k_loss.append(history.history['loss'])
        k_loss_test.append(history.history['val_loss'])
    
    avg_time=np.mean(time_log)
    return k_accuracy, k_accuracy_test, k_loss, k_loss_test, avg_time
```

```python
# Plot time against batch size
def plot_time(batch_sizes, avg_time):
    plt.clf()
    plt.plot(batch_sizes, avg_time)
    plt.ylabel('Average time/epoch')
    plt.xlabel('Batch size')
    plt.xticks(batch_sizes)
    plt.title('Average time/epoch for different batch sizes')
```

```python
# Main code
batch_sizes = [4,8,16,32,64]
avg_time_logs = []
acc_log = []
test_acc_log = []
loss_log = []
val_loss_log = []
epochs = 300

TrainX,TrainY=get_data()

# running 5 fold cross validation for different batch sizes
for x in batch_sizes:
    print("Running model for batch size: ",x)
    accuracy, val_acc, loss, val_loss, avg_time = k_fold_CV(5, TrainX, TrainY, epochs, x)
    
    acc_log.append([np.mean([y[i] for y in accuracy]) for i in range(epochs)])
    test_acc_log.append([np.mean([y[i] for y in val_acc]) for i in range(epochs)])
    loss_log.append([np.mean([y[i] for y in loss]) for i in range(epochs)])
    val_loss_log.append([np.mean([y[i] for y in val_loss]) for i in range(epochs)])
    
    avg_time_logs.append(avg_time)
    #plot_accuracy(accuracy, val_acc,epochs,x)
    #plt.savefig('Question1B_'+str(x)+'.png')
```
