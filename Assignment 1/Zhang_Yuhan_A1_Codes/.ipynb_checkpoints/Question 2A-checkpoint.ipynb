```python
#
# Project 1, source code for part B
#

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import time
```

```python
# scale data (set range min and max of 0 and 1 respectively)
def scale(X, X_min, X_max):
    return (X - X_min)/(X_max-X_min)

# Default values
NUM_CLASSES = 7

epochs = 1000
batch_size = 8

decay_beta = 10**-3
learning_rate = 10**-3
hidden_layer = 10

seed = 10
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#read train data
def get_data():
    train_input = np.genfromtxt('admission_predict.csv', delimiter= ',')
    trainX, train_Y = train_input[1:,1:8], train_input[1:,-1]
    trainX = scale(trainX, np.min(trainX, axis=0), np.max(trainX, axis=0))
    trainY = train_Y.reshape(train_Y.shape[0], 1)
    
    idx = np.arange(trainX.shape[0])
    np.random.shuffle(idx)
    trainX, trainY = trainX[idx], trainY[idx]
    
    return trainX, trainY

#Split train and test data (70:30)
def split_data(trainX, trainY):
    train_X, test_X, train_Y, test_Y = train_test_split(trainX, trainY, test_size=0.3)
    train_X = (train_X- np.mean(train_X, axis=0))/ np.std(train_X, axis=0)
    print("TrainX: ",train_X.shape)
    print("TestX: ",test_X.shape)
    print("TrainY: ",train_Y.shape)
    print("TestY: ",test_Y.shape)
    return train_X, test_X, train_Y, test_Y
```

```python
#model
def get_model():
    model = keras.Sequential([
        keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
        keras.layers.Dense(1)
    ])

    SGD_opt = keras.optimizers.SGD(learning_rate=learning_rate)

    model.compile(optimizer=SGD_opt,
                  loss=keras.losses.MeanSquaredError(),
                  metrics=['mse'])
    return model
```

```python
def fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, isCallback):
    if isCallback:
        # stops training when there is no observed improvement in mse for 5 epochs
        callback = tf.keras.callbacks.EarlyStopping(monitor='val_mse', min_delta = 0.0001,patience=5)
    else:
        callback = timeCallback()
    history = model.fit(train_X, train_Y,
                        epochs=epochs,
                        callbacks = [callback],
                        verbose = 2,
                        batch_size=batch_size,
                        validation_data = (test_X, test_Y),
                        shuffle=True)
    #timeTaken = np.mean(timeTaken.logs)
    return history, model
```

```python
def plot_mse(mse, mse_test, epochs):
    plt.clf()
    plt.plot(range(epochs), mse)
    plt.plot(range(epochs), mse_test)
    plt.xlabel('No. epochs')
    plt.ylabel('mse')
    plt.legend(['train mse', 'test mse'])
    plt.title('mse of train and test data')
    
#plot_accuracy(accuracy,val_acc,100,64)
```

```python
class timeCallback(tf.keras.callbacks.Callback):
    
    def on_train_begin(self, logs={}):
        self.logs=[]

    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_start = time.time()

    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(time.time() - self.epoch_start)
```

```python
# Without early stopping
TrainX, TrainY = get_data()
train_X, test_X, train_Y, test_Y = split_data(TrainX, TrainY)
model = get_model()
history, model = fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, False)
```

```python
# Mse of training and testing data without early stopping
plot_mse(history.history['mse'], history.history['val_mse'], len(history.history['mse']))
```

```python
# With early stopping
TrainX, TrainY = get_data()
train_X, test_X, train_Y, test_Y = split_data(TrainX, TrainY)
model = get_model()
history, model = fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, True)
```

```python
plot_mse(history.history['mse'], history.history['val_mse'], len(history.history['mse']))
```

```python
testY_50 = test_Y[:50]
testX_50 = test_X[:50]
predictedY_50 = model.predict(x=testX_50)
print('root mse: ', np.sqrt(mean_squared_error(testY_50, predictedY_50)))
```

```python
plt.figure()
plt.plot(testY_50)
plt.plot(predictedY_50)
plt.legend(['Actual Value', 'Predicted Value',], loc='lower left')
plt.title('Actual and Predicted values for 50 test samples')
```
