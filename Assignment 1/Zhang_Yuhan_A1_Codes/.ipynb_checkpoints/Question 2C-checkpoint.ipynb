```python
#
# Project 1, source code for part b
#

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import time
```

```python
# scale data (set range min and max of 0 and 1 respectively)
def scale(X, X_min, X_max):
    return (X - X_min)/(X_max-X_min)

# Default values
NUM_CLASSES = 7

epochs = 300
batch_size = 64

decay_beta = 10**-3
learning_rate = 10**-3
hidden_layer = 50

seed = 10
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#read train data
def get_data():
    train_input = np.genfromtxt('admission_predict.csv', delimiter= ',')
    trainX, train_Y = train_input[1:,1:8], train_input[1:,-1]
    trainX = scale(trainX, np.min(trainX, axis=0), np.max(trainX, axis=0))
    trainY = train_Y.reshape(train_Y.shape[0], 1)
    
    idx = np.arange(trainX.shape[0])
    np.random.shuffle(idx)
    trainX, trainY = trainX[idx], trainY[idx]
    
    return trainX, trainY

#Split train and test data (70:30)
def split_data(trainX, trainY):
    train_X, test_X, train_Y, test_Y = train_test_split(trainX, trainY, test_size=0.3)
    train_X = (train_X- np.mean(train_X, axis=0))/ np.std(train_X, axis=0)
    print("TrainX: ",train_X.shape)
    print("TestX: ",test_X.shape)
    print("TrainY: ",train_Y.shape)
    print("TestY: ",test_Y.shape)
    return train_X, test_X, train_Y, test_Y
```

```python
#model
def get_model_3():
    model = keras.Sequential([
        keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
        keras.layers.Dense(1)
    ])

    SGD_opt = keras.optimizers.SGD(learning_rate=learning_rate)

    model.compile(optimizer=SGD_opt,
                  loss=keras.losses.MeanSquaredError(),
                  metrics=['mse'])
    return model

def get_model_4(isDropout):
    if isDropout:
        model = keras.Sequential([
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dropout(rate=0.2, seed=seed),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dropout(rate=0.2, seed=seed),
            keras.layers.Dense(1)
        ])
    else:
        model = keras.Sequential([
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dense(1)
        ])

    SGD_opt = keras.optimizers.SGD(learning_rate=learning_rate)

    model.compile(optimizer=SGD_opt,
                  loss=keras.losses.MeanSquaredError(),
                  metrics=['mse'])
    return model

def get_model_5(isDropout):
    if isDropout:
        model = keras.Sequential([
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dropout(rate=0.2, seed=seed),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dropout(rate=0.2, seed=seed),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dropout(rate=0.2, seed=seed),
            keras.layers.Dense(1)
        ])
    else:
        model = keras.Sequential([
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dense(hidden_layer, use_bias = True, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(decay_beta)),
            keras.layers.Dense(1)
        ])

    SGD_opt = keras.optimizers.SGD(learning_rate=learning_rate)

    model.compile(optimizer=SGD_opt,
                  loss=keras.losses.MeanSquaredError(),
                  metrics=['mse'])
    return model
```

```python
def fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, isCallback):
    if isCallback:
        # stops training when there is no observed improvement in mse for 5 epochs
        callback = tf.keras.callbacks.EarlyStopping(monitor='mse', min_delta = 0.0001,patience=5)
    else:
        callback = timeCallback()
    history = model.fit(train_X, train_Y,
                        epochs=epochs,
                        callbacks = [callback],
                        verbose = 2,
                        batch_size=batch_size,
                        validation_data = (test_X, test_Y),
                        shuffle=True)
    #timeTaken = np.mean(timeTaken.logs)
    return history, model
```

```python
def plot_mse(mse, mse_test, epochs):
    plt.clf()
    plt.plot(range(epochs), mse)
    plt.plot(range(epochs), mse_test)
    plt.xlabel('No. epochs')
    plt.ylabel('mse')
    plt.legend(['train mse', 'test mse'], loc='lower right')
    plt.title('mse train and test data')
    
#plot_accuracy(accuracy,val_acc,100,64)
```

```python
class timeCallback(tf.keras.callbacks.Callback):
    
    def on_train_begin(self, logs={}):
        self.logs=[]

    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_start = time.time()

    def on_epoch_end(self, epoch, logs={}):
        self.logs.append(time.time() - self.epoch_start)
```

```python
# Model with all features
TrainX, TrainY = get_data()
train_X, test_X, train_Y, test_Y = split_data(TrainX, TrainY)
model = get_model()
history_original, model = fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, False)
plot_mse(history_original.history['mse'], history_original.history['val_mse'], len(history_original.history['mse']))
```

```python
def recursive_drop(original_mse, train_X, test_X, train_Y, test_Y, keepTrack):
    mse_log = []
    test_mse_log = []
    improved_mse = []

    # Delete one feature from original dataset
    for i in range(len(train_X[0])):
        print("Dropping feature at index: ",i)
        dropped_trainX = np.delete(train_X,0,1)
        dropped_testX = np.delete(test_X,0,1)

        model = get_model()
        history, model = fit_model(dropped_trainX, train_Y, dropped_testX, test_Y, epochs, batch_size, model, False)
        mse_log.append(history.history['mse'])
        test_mse_log.append(history.history['val_mse'])
    
    for i in range(len(mse_log)):
        print('Checking new mse against original: ', i)
        if str(mse_log[i][-1]) < str(original_mse[-1]):
            improved_mse.append(mse_log[i][-1])
            print("Original mse: \t",original_mse[-1])
            print("Improved mse: \t",mse_log[i][-1])
    
    if len(improved_mse):
        drop_index = improved_mse.index(min(improved_mse))
        new_trainX = np.delete(train_X,drop_index,1)
        new_testX = np.delete(test_X,drop_index,1)
        print("\nDropped feature at index: ", drop_index)
        if keepTrack:
            order.append(drop_index)
        return recursive_drop(improved_mse[drop_index], new_trainX, new_testX, train_Y, test_Y, keepTrack)
    elif len(train_X[0]) == 2:
        return mse_log[0], test_mse_log[0]
    else:
        return mse_log[0], test_mse_log[0]
```

```python
def normal_drop():
    TrainX, TrainY = get_data()
    train_X, test_X, train_Y, test_Y = split_data(TrainX, TrainY)
    mse_log = []
    test_mse_log = []

    model = get_model()
    history, model = fit_model(train_X, train_Y, test_X, test_Y, epochs, batch_size, model, False)
    mse_log.append(history.history['mse'])
    test_mse_log.append(history.history['val_mse'])

    # Delete one feature from original dataset
    for i in range(len(TrainX[0])):
        print("Dropping feature at index: ",i)
        dropped_trainX = np.delete(train_X,0,1)
        dropped_testX = np.delete(test_X,0,1)

        model = get_model()
        history, model = fit_model(dropped_trainX, train_Y, dropped_testX, test_Y, epochs, batch_size, model, False)
        mse_log.append(history.history['mse'])
        test_mse_log.append(history.history['val_mse'])
    return mse_log, test_mse_log
```
